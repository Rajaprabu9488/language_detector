```python
import re

def remove_sub(text):
    text = re.sub(r'[!@#$(),n"%^*?:;~`0-9\[\]]', ' ', text)
    return text
```

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
test_sample=encoder.fit_transform(df['Language'])
test_sample.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(analyzer="char",
    ngram_range=(3,5),
    min_df=5,
    max_features=100_000,
    lowercase=True)
train_sample=tfidf.fit_transform(df['Text'])
# train_sample=train_sample.toarray()
train_sample.shape
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(train_sample,test_sample,test_size=0.2,stratify=test_sample)

print(xtrain.shape)
print(xtest.shape)
```

```python
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
```

```python
class TextDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return self.X.shape[0]  

    def __getitem__(self, idx):
        return (
            torch.tensor(self.X[idx].toarray(), dtype=torch.float32).squeeze(0),
            torch.tensor(self.y[idx], dtype=torch.long)
        )
```

```python
batch_size = 32

train_dataset = TextDataset(xtrain, ytrain)
test_dataset  = TextDataset(xtest, ytest)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True,pin_memory=True)
test_loader  = DataLoader(test_dataset, batch_size=batch_size,pin_memory=True)
```

```python
train_loader
```

```python
class model_v0(nn.Module):
    def __init__(self,input,output):
        super().__init__()
        self.layer=nn.Sequential(
            nn.Linear(in_features=input,out_features=128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(in_features=128,out_features=output),
        )
    def forward(self,x):
        return self.layer(x)
```

```python
model_0=model_v0(xtrain.shape[1],17)
model_0.state_dict()
```

```python
loss_fn=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model_0.parameters(),lr=0.01)

```

```python
model_0.to(device="cuda")
```

```python
def accuracy_score(prediction,label):
    result=(torch.argmax(prediction)==label).sum().item()
    return result
```

```python
epochs=20
train_losses=[]
test_losses=[]
total_train_acc=[]
total_test_acc=[]

for epoch in range(epochs):
    model_0.train()
    total_train=0
    train_loss=0
    train_acc=0
    for X,Y in train_loader:
        X=X.to("cuda")
        Y=Y.to("cuda")
        preds=model_0(X)
        loss=loss_fn(preds,Y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        train_loss+=loss.item()
        total_train+=Y.size(0)
        train_acc+=accuracy_score(preds,Y)
    train_losses.append(train_loss/len(train_loader))
    total_train_acc.append((train_acc/total_train)*100)
    model_0.eval()
    test_loss=0
    total_test=0
    test_acc=0
    with torch.inference_mode():
        for X,Y in test_loader:
            X=X.to("cuda")
            Y=Y.to("cuda")
            test_preds=model_0(X)
            loss=loss_fn(test_preds,Y)
            test_loss+=loss.item()
            total_test+=Y.size(0)
            test_acc+=accuracy_score(test_preds,Y)
        test_losses.append(test_loss/len(test_loader))
        total_test_acc.append((test_acc/total_test)*100)
    print(f"epochs: {epoch} | train loss: {train_losses[-1]} | test loss: {test_losses[-1]} | train accuracy: {total_train_acc[-1]} | test accuracy: {total_test_acc[-1]}")
    
```

```python
lang=["Learning is a lifelong journey.",
      "सीखना एक आजीवन यात्रा है।",
      "கற்பது வாழ்நாள் முழுவதும் தொடரும் ஒரு பயணம்.",
      "ಕಲಿಕೆ ಜೀವನ ಪರ್ಯಂತದ ಒಂದು ಪ್ರಯಾಣ.",
      "Aprender es un viaje de toda la vida.",
      "Hoe gaat het met je?"]
def predict_language(model,lang,text_transformer,label_encoder):
    for text in lang:
        model.eval()
        with torch.inference_mode():
            transform_text=remove_sub(text)
            transform_text=text_transformer.transform([transform_text.lower()])
            transform_text=transform_text.toarray()
            transform_text=torch.tensor(transform_text,dtype=torch.float,device="cuda")
            bb=model(transform_text)
            result=torch.argmax(bb,dim=1)
            int_to_lang=label_encoder.inverse_transform(result.cpu())
            print(f"{text} is from {int_to_lang}")
predict_language(model_0,lang,tfidf,encoder)
```

```python
from pathlib import Path
import joblib
path=Path("./model_0")
path.mkdir(exist_ok=True)
model_path=path/"model_0.pth"
encoder_path=path/"label_encoder.pkl"
tfidf_path=path/"tfidf_vector.pkl"

print(f"model_0 saving in {model_path}...")
torch.save(obj=model_0.state_dict(),f=model_path)
print("...file saved...")

joblib.dump(tfidf,tfidf_path)
joblib.dump(encoder,encoder_path)
```
